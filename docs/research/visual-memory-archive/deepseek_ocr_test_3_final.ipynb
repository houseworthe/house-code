## Cell 5: Load Model & Run OCR (~10-30 sec inference)

**FIXED**: Using official DeepSeek-OCR parameters from Context7 docs

````python
from transformers import AutoModel, AutoTokenizer
import torch
import time
import os

# CRITICAL: Create output directory first
os.makedirs('/content/ocr_output', exist_ok=True)

print("Loading DeepSeek-OCR model...")
tokenizer = AutoTokenizer.from_pretrained(
    '/content/deepseek-ocr',
    trust_remote_code=True
)

model = AutoModel.from_pretrained(
    '/content/deepseek-ocr',
    _attn_implementation='flash_attention_2',
    trust_remote_code=True,
    use_safetensors=True  # Add this!
).eval().cuda().to(torch.bfloat16)

print("✓ Model loaded to GPU\n")
print("Running OCR inference...\n")

start = time.time()

# Use exact parameters from official docs
result = model.infer(
    tokenizer,
    prompt="<image>\nFree OCR.",
    image_file='/content/test_image.png',
    output_path='/content/ocr_output',  # Must be valid directory
    base_size=1024,
    image_size=640,
    crop_mode=True,
    save_results=True,
    test_compress=True
)

elapsed = time.time() - start
print(f"\n✓ OCR complete in {elapsed:.2f}s")
print(f"✓ Output: {len(result)} characters")
````

````python
from transformers import AutoModel, AutoTokenizer
import torch
import time
import os

# CRITICAL: Create output directory first
os.makedirs('/content/ocr_output_grounding', exist_ok=True)

print("Loading DeepSeek-OCR model...")
tokenizer = AutoTokenizer.from_pretrained(
    '/content/deepseek-ocr',
    trust_remote_code=True
)

model = AutoModel.from_pretrained(
    '/content/deepseek-ocr',
    _attn_implementation='flash_attention_2',
    trust_remote_code=True,
    use_safetensors=True
).eval().cuda().to(torch.bfloat16)

print("✓ Model loaded to GPU\n")
print("Running OCR inference with GROUNDING prompt...\n")

start = time.time()

# TEST GROUNDING PROMPT FOR CODE
result = model.infer(
    tokenizer,
    prompt="<image>\nExtract all text preserving exact indentation and spacing.",    image_file='/content/test_image.png',
    output_path='/content/ocr_output_grounding',
    base_size=1024,
    image_size=640,
    crop_mode=True,
    save_results=True,
    test_compress=True
)

# Load result from file
with open('/content/ocr_output_grounding/result.mmd', 'r') as f:
    result = f.read()

elapsed = time.time() - start
print(f"\n✓ OCR complete in {elapsed:.2f}s")
print(f"✓ Output: {len(result)} characters")
````

## Cell 6: Calculate Accuracy & Display Results

Character-level accuracy using difflib.

````python
import difflib

# Calculate accuracy
matcher = difflib.SequenceMatcher(None, ORIGINAL_TEXT.strip(), result.strip())
accuracy = matcher.ratio() * 100

# Display results
print("\n" + "="*70)
print("🎯 PHASE 2 RESULTS")
print("="*70)
print(f"CHARACTER ACCURACY: {accuracy:.2f}%")
print(f"LATENCY: {elapsed:.2f}s")
print(f"ORIGINAL: {len(ORIGINAL_TEXT)} chars")
print(f"OCR OUTPUT: {len(result)} chars")
print(f"DIFFERENCE: {abs(len(result) - len(ORIGINAL_TEXT))} chars")
print("="*70)

# Token compression estimate
text_tokens = len(ORIGINAL_TEXT) // 4
visual_tokens = 256 * 2  # 2 pages at 1024x1024 = 256 tokens each
token_compression = text_tokens / visual_tokens
print(f"\nEst. text tokens: ~{text_tokens}")
print(f"Visual tokens (2 pages): {visual_tokens}")
print(f"Token compression: {token_compression:.2f}x")
````
